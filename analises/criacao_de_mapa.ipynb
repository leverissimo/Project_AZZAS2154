```python
import json
import folium
from geopy.geocoders import Photon
import folium
import time


path = "dados/filiais_UF_RJ.json"

with open(path, "r", encoding="utf-8") as f:
    dados = json.load(f)

# --- Detectar automaticamente o formato ---
if isinstance(dados, list):
    # Formato 2: já é lista
    lista_filiais = dados
elif isinstance(dados, dict):
    # Formato 1: pegar a primeira chave e acessar a lista
    chave_principal = list(dados.keys())[0]
    lista_filiais = dados[chave_principal]
else:
    raise ValueError("Formato de JSON não reconhecido")
```

```python
geolocator = Photon(user_agent="mapa_filiais")

ufs = {filial["UF"] for filial in lista_filiais if "UF" in filial}

if len(ufs) == 1:
    uf = list(ufs)[0]
    local_uf = geolocator.geocode(f"{uf}, Brasil")
    if local_uf:
        centro = [local_uf.latitude, local_uf.longitude]
    else:
        centro = [-23.55, -46.63]  # fallback São Paulo
else:
    # várias UFs → centraliza no Brasil
    centro = [-14.235, -51.925]

# Criar mapa centralizado dinamicamente
mapa = folium.Map(location=centro, zoom_start=8)

for filial in lista_filiais:
    endereco_completo = f"{filial['ENDERECO']}, {filial['BAIRRO']}, {filial['CIDADE']}, {filial['UF']}, Brasil"
    print(endereco_completo)
    try:
        local = geolocator.geocode(endereco_completo, timeout=10)
        if local:
            folium.Marker(
                [local.latitude, local.longitude],
                popup=f"{filial['FILIAL']} ({filial['COD_FILIAL']})",
                tooltip=filial['FILIAL']
            ).add_to(mapa)
        else:
            print(f"Não encontrado: {filial['FILIAL']} → {endereco_completo}")
    except Exception as e:
        print(f"Erro ao localizar {filial['FILIAL']}: {e}")

    time.sleep(1)  # respeita limite do serviço

mapa.save("mapa_filiais.html")
```
